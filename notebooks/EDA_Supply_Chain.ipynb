## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1 Import required libraries
import pandas as pd
import numpy as np

# No longer creating a separate copy if we want to clean df directly for consistent use
# df_clean = df.copy()

# Convert 'Weight (Kilograms)' and 'Freight Cost (USD)' to numeric, coercing errors
# This is crucial as these columns were identified as 'object' type but contain numerical data
df['Weight (Kilograms)'] = pd.to_numeric(df['Weight (Kilograms)'], errors='coerce')
df['Freight Cost (USD)'] = pd.to_numeric(df['Freight Cost (USD)'], errors='coerce')

# 3 Check missing values
# df.isnull().sum() # This was executed, kept for reference

# 4 Handle missing values

# Fill categorical columns with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Fill numerical columns with median
# This will now correctly handle NaNs introduced by pd.to_numeric for the converted columns
for col in df.select_dtypes(include=['int64','float64']).columns:
    df[col].fillna(df[col].median(), inplace=True)

# 5 Check duplicate rows
# df.duplicated().sum() # This was executed, kept for reference

# Remove duplicates
df.drop_duplicates(inplace=True)

# 6 Convert data types if needed
df['ID'] = df['ID'].astype(int)

# 7 Check dataset information after cleaning
df.info()

# 8 Verify dataset shape
df.shape
````

### What all manipulations have you done and insights you found?

The dataset underwent several preprocessing steps to ensure data quality and consistency. Key numerical columns such as Freight Cost and Weight were converted from object to numeric types to enable analysis. Missing values were handled using mode for categorical variables and median for numerical variables to maintain robustness against outliers. Duplicate records were removed to ensure data integrity. These steps resulted in a clean and structured dataset suitable for exploratory data analysis and further modeling.
